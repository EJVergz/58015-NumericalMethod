```
import numpy as np
import matplotlib.pyplot as plt
import math
```

#Secant Method


*Â© A.D Buenaventura, M.A Catinoy and E.J Vergara 2021*

```
def f(x): return 2*x**2 - 4*x + 1
x0, x1 = 2, 3
epochs = 100
root = None
value_x = []
value_x_prime = []
for epoch in range(epochs):
  x2 = x1 - (f(x1)*(x1-x0))/(f(x1)-f(x0))
  value_x.append(x1)
  value_x_prime.append(x2)

  if np.allclose(x1, x2): 
    root = x2
    break
  else:
    x0,x1 = x1, x2
print("The root is {:.04f}, found at {} epochs".format(root, epoch))

value_x = value_x[0:]
value_x_prime = value_x_prime[0:]
x_range = np.arange(len(value_x),dtype=int)
plt.figure(figsize=(10,7))
plt.plot(x_range, value_x, 'o--', color = 'blue', label='x')
plt.plot(x_range, value_x_prime, 'D--', color = 'green', label='x')
plt.xlabel("epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1,1.0))
plt.legend()
plt.grid()
plt.show()
```

#Regula Falsi

```
def f(x): return 2*x**2 - 5*x - 1
x0, x1 = 1.1,3
y1, y2 = f(x0), f(x1)
root = None 
pos = 0
value_x = []
value_x_prime = []
if np.allclose(0,y1): root = x0
elif np.allclose(0,y2): root = x1
elif np.sign(y1) == np.sign(y2):
  print("No root here")
else:
  for pos in range(0,100):
    x2 = x1 - (f(x1)*(x1-x0))/(f(x1)-f(x0))
    value_x.append(x1)
    value_x_prime.append(x2)
    if np.allclose(0,f(x2), 1e-06):
      root = x2
      break
    if np.sign(f(x0)) != np.sign(f(x2)): x1,y2 = x2,f(x2)
    else: x0,y1 = x2,f(x1) 

print("The root is {:.04f}, found at {} epoch".format(root, pos))

value_x = value_x[0:]
value_x_prime = value_x_prime[0:]
x_range = np.arange(len(value_x),dtype=int)
plt.figure(figsize=(10,7))
plt.plot(x_range, value_x, 'o--', color = 'dodgerblue', label='x')
plt.plot(x_range, value_x_prime, 'D--', color = 'orange', label='x')
plt.xlabel("epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1,1.0))
plt.legend()
plt.grid()
plt.show()
```
