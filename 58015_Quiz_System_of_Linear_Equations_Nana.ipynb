Suppose [1] you want to make a certain kind of tropical punch, using bananas, oranges, and papayas. You don't know how many of each to put in the punch, but you know that there are seven pieces of fruit in the mix, and there are twice as many oranges as bananas. You also know that the seven pieces of fruit cost 5.25, where bananas cost 0.50 each, oranges cost 0.75 each, and papayas cost 0.25 each.

#Standard Bracketed Form
\begin{array}{l}\text{                    }b + o + p = 7\hfill \\ \text{                      }0.5b + 0.75o + 1.25p = 5.25\hfill \\ o = 2b\hfill \end{array}


#Linear Combination Form (Matrix)
$$\begin{bmatrix}
1&1&1\\
0.5&0.75&1.25\\
-2&1&0\\
\end{bmatrix} \cdot
\begin{bmatrix}x_1\\x_2\\x_3\\\end{bmatrix}=
\begin{bmatrix}7\\5.25\\0\\\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
A = np.array([[1, 1, 1], [0.5, 0.75, 1.25], [-2, 1, 0]])
B = np.array([7, 5.25, 0])
X = np.linalg.solve(A,B)

print(X)
```

In the line 3&4, we use np.array to put the equation that is needed to be solved. We used np.linalg.solve(A, B) to solve a linear equation, it computes the exact solution.

#Vector Visualization


```python
fig = plt.figure()
ax = plt.axes(projection="3d")
Eq_1 = [1, 1, 1] 
Eq_2 = [0.5, 0.75, 1.25]
Eq_3 = [-2, 1, 0]
ax.plot3D(Eq_1, Eq_2, Eq_3)
plt.show()
```

We use ax = plt.axes(projection="3d") because it functions as a container which we can view the 3D results by using ax.plot3D(Eq_1, Eq_2, Eq_3) where the 3 variables have their own triple ordered pair and was computed by the program, then as shown above the results formed a 3D figure. This can be used in more complex and identical figures or even make images in 3D.
