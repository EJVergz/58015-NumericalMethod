```python
!pip install pydataset
from pydataset import data
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from matplotlib import pyplot as plt
```

```python
pima = data('Pima.tr')
X_train, X_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
plt.scatter(X_train, y_train, label = 'Training Data', color = 'r', alpha = .7)
plt.scatter(X_test, y_test, label = 'Testing Data', color = 'g', alpha = .7)
plt.legend()
plt.title("Test Train Split")
plt.show()
```

#Evaluation

```python
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
prediction = LR.predict(X_test.values.reshape(-1,1))
plt.plot(X_test, prediction, label = 'Linear Regression', color = 'b')
plt.scatter(X_test, y_test, label = 'Actual Test Data', color = 'g', alpha = .7)
plt.legend()
plt.show()
```

#**System of Linear Equation**
*by Â© A.D Buenaventura, M.A Catinoy, and E.J Vergara*

John  invested his inheritance of 12,000 in three different funds: part in a money-market fund paying 3% interest annually; part in municipal bonds paying 4% annually; and the rest in mutual funds paying 7% annually. John invested 4,000 more in mutual funds than he invested in municipal bonds. The total interest earned in one year was 670. How much did he invest in each type of fund? [[1]](https://courses.lumenlearning.com/wmopen-collegealgebra/chapter/introduction-systems-of-linear-equations-three-variables/)


#Standard Bracketed Form
\begin{array}{l}\text{                    }x+y+z=12,000\hfill \\ \text{                      }-y+z=4,000\hfill \\ 0.03x+0.04y+0.07z=670\hfill \end{array}
To make the calculations simpler, we can multiply the third equation by 100. Thus,

\begin{array}{ll}\text{  }x+\text{   }y+z\text{ }=12,000\hfill & \hfill \\ \text{       }-y+z\text{ }=4,000\hfill & \hfill \\ 3x+4y+7z=67,000\hfill & \hfill \end{array}

#Linear Combination Form (Matrix)
$$\begin{bmatrix}
1&1&1\\
0&-1&1\\
3&4&7\\
\end{bmatrix} \cdot
\begin{bmatrix}x_1\\x_2\\x_3\\\end{bmatrix}=
\begin{bmatrix}12,000\\4,000\\67,000\\\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
A = np.array([[1, 1, 1], [0, -1, 1], [3, 4, 7]])
B = np.array([12000, 4000, 67000])
X = np.linalg.solve(A,B)

print(X)
```

In the line 3&4, we use np.array to put the equation that is needed to be solved. We used np.linalg.solve(A, B) to solve a linear equation, it computes the exact solution.

#Vector Visualization


```python
fig = plt.figure()
ax = plt.axes(projection="3d")
X = [1, 1, 1] 
Y = [0, -1, 1]
Z = [3, 4, 7]
ax.plot3D(X, Y, Z)
plt.show()
```

We use ax = plt.axes(projection="3d") because it functions as a container which we can view the 3D results by using ax.plot3D(X, Y, Z) where the 3 variables have their own triple ordered pair and was computed by the program, then as shown above the results formed a 3D figure. This can be used in more complex and identical figures or even make images in 3D.
